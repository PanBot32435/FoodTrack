```python
# Import necessary libraries
import pandas as pd

# Load the dataset from CSV file
df = pd.read_csv("onlinefoods.csv")

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Check the structure and data types of the columns
print("\nData types of the columns:")
print(df.dtypes)

# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Basic statistics summary
print("\nBasic statistics summary:")
print(df.describe())

# Visualize distribution of numerical variables (e.g., age, monthly income)
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Bar plot for gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Box plot for monthly income distribution
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Monthly Income', color='salmon')
plt.title('Distribution of Monthly Income')
plt.ylabel('Monthly Income')
plt.show()

# Scatter plot for latitude and longitude
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='longitude', y='latitude', color='green')
plt.title('Geographical Distribution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
